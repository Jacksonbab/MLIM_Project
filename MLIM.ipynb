```python
import os
from PIL import Image
from google.colab import drive
drive.mount('/content/drive')
base_dir = "/content/drive/MyDrive/sysu-shape-dataset-master/sysu-shape-dataset-master"  # Replace with the path to your folders
new_base_dir = "/content/drive/MyDrive/MLIM_dataset"
folders = ["airplane", "bicycle", "boat", "car", "motorbike"]
target_size = (256, 256)  # Target image size

def resize_images_in_folder(category_folder, new_category_folder):
    images_path = os.path.join(category_folder, "images")
    new_images_path = os.path.join(new_category_folder, "images")
    # Ensure the new directory structure exists
    os.makedirs(new_images_path, exist_ok=True)

    if not os.path.exists(images_path):
        print(f"No 'images' folder found in {category_folder}. Skipping...")
        return

    for file in os.listdir(images_path):
        if file.lower().endswith('.jpg'):  # Process only .jpg files
            file_path = os.path.join(images_path, file)
            new_file_path = os.path.join(new_images_path, file)  # New file path
            try:
                with Image.open(file_path) as img:
                    img = img.resize(target_size, Image.Resampling.LANCZOS)  # Resize the image
                    img.save(new_file_path)  # Save the resized image in the new directory
                print(f"Resized and saved: {new_file_path}")
            except Exception as e:
                print(f"Error resizing {file_path}: {e}")

# Iterate through each category folder
for folder in folders:
    category_folder = os.path.join(base_dir, folder)
    new_category_folder = os.path.join(new_base_dir, folder)
    if os.path.exists(category_folder):
        resize_images_in_folder(category_folder, new_category_folder)
    else:
        print(f"Category folder not found: {category_folder}")
```
